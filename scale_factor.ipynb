```python
import numpy as np
import pylab
import astropy.io.fits as pyfits
from scipy.ndimage import gaussian_filter

# Others will need to change this path...
ROOT = '/Users/maryame/Desktop/Data_1373/'

FILES = [
    'jw01373010001_03103_00001_nrcblong/jw01373010001_03103_00001_nrcb4_cal.fits',
]

# Read arrays
arrays = []
for file in FILES:
    hdulist = pyfits.open(ROOT + file)
    arrays.append(hdulist[1].data)

raw_image1 = np.array(arrays)
print(raw_image1.shape)
```

```python
def myshow(im, vmin=None, vmax=None, **keywords):
    """Function to show one image at reasonable size."""
    vmin = vmin or -500    # an OK default
    vmax = vmax or 1000    # an OK default
    if 'origin' not in keywords:
        keywords['origin'] = 'lower'
    pylab.figure(figsize=(6,6), dpi=300)
    pylab.gray()
    pylab.imshow(im, vmin=vmin, vmax=vmax, **keywords)

def flatshow(im, vmin=None, vmax=None, **keywords):
    """Quick function to high-pass filter and then show the result."""
    im = im - gaussian_filter(im,20)
    myshow(im, vmin=vmin, vmax=vmax, **keywords)
```

```python
myshow(raw_image1[0], vmin=0, vmax=50000) # Obs 10, first dither, first integration, first group

# This is an un-processed image
```
